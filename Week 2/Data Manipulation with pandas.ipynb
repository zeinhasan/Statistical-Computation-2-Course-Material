## 2.1 Efficient Summaries

```python
"""The .agg() method allows you to apply your own custom functions to a DataFrame, 
as well as apply functions to more than one column of a DataFrame at once, 
making your aggregations super-efficient. For example"""

# A custom IQR function
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)
    
# Print IQR of the temperature_c column
sales["temperature_c"].agg(iqr)
```

```python
# Update to print IQR of temperature_c, fuel_price_usd_per_l, & unemployment
sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg(iqr)
```

```python
# Update to print IQR and median of temperature_c, fuel_price_usd_per_l, & unemployment
sales[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr,np.median])
```

## 2.2 Cumulative statistics

```python
"""Cumulative statistics can also be helpful in tracking summary statistics over time. 
In this exercise, you'll calculate the cumulative sum and cumulative max of a department's weekly sales, 
which will allow you to identify what the total sales were so far as well as what the highest weekly sales were so far."""

# Sort sales_1_1 by date
sales_1_1 = sales.sort_values("date")

# Get the cumulative sum of weekly_sales, add as cum_weekly_sales col
sales_1_1["cum_weekly_sales"] = sales_1_1["weekly_sales"].cumsum()

# Get the cumulative max of weekly_sales, add as cum_max_sales col
sales_1_1["cum_max_sales"] = sales_1_1["weekly_sales"].cummax()

# Check the result
sales_1_1
```

## 2.4 Groupby

```python
"""you'll calculate the total sales made at each store type, without using .groupby(). 
You can then use these numbers to see what proportion of Walmart's total sales were made at each type."""

# Calc total weekly sales
sales_all = sales["weekly_sales"].sum()

# Subset for type A stores, calc total weekly sales
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()

# Subset for type B stores, calc total weekly sales
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()

# Get proportion for each type
sales_propn_by_type = [sales_A, sales_B] / sales_all

# Print sales_propn_by_type
sales_propn_by_type
```

```python
# you'll calculate the total sales made at each store type, using .groupby(). 
# Group by type; calc total weekly sales
sales_by_type = sales.groupby("type")["weekly_sales"].sum()

# Get proportion for each type
sales_propn_by_type = sales_by_type / sum(sales_by_type)

# Print sales_propn_by_type
sales_propn_by_type
```

```python
# For each store type, aggregate weekly_sales: get min, max, mean, and median
sales_stats = sales.groupby("type")["weekly_sales"].agg([np.min, np.max, np.mean, np.median])

# Print sales_stats
sales_stats
```

```python
# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median
unemp_fuel_stats = sales.groupby("type")[["unemployment", "fuel_price_usd_per_l"]].agg([np.min, np.max, np.mean, np.median])

# Print unemp_fuel_stats
unemp_fuel_stats
```

# 3. Slicing and Indexing DataFrames

## 3.1 Setting and Removing Index

What we call 'index' is the leftmost column, usually it starts at 0 and ascending.

On many occasions, we may want to **adjust the index to better identify our data** (for example, we have Temperature data recorded every day, then it would make sense to replace the index with **observation date**).

In this section, we will use temperatures dataset. This dataset represents the daily average temperature in cities across many nations. It consists of four columns, which are observation date, city, country, and the average temperature in degree Celcius.

```python
temperatures
```

We can set a certain column to be the index. This way we can easily refer all rows by the city name. 

Please note that each row doesn't have to have different index. 

```python
# Index temperatures by city
temperatures_ind = temperatures.set_index("city")

# Look at temperatures_ind
temperatures_ind
```

We notice that there are multiple rows with the same index! What if we try to access the row with index = 'Xian' ?

```python
temperatures_ind.loc['Xian',:]
```

It returns ALL rows with the city 'Xian'! This is very useful in certain cases when we just want to aggregate our data (especially in univariate time series analysis).

It is also possible to reset the index to the original index. Simply call the 'reset_index' method and you're good to go!

```python
# Reset the index, keeping its contents
temperatures_ind.reset_index(inplace=True)

# Check the result 
temperatures_ind
```

## 3.2 Subsetting with .loc[]

Let's say we want to select some cities from our dataset. There are many ways to do it, some are perhaps easier to remember than the other ;)

```python
# Make a list of cities to subset on
cities = ["Moscow", "Saint Petersburg"]
```

One common way to do this is by using square brackets. 

Recall that in order to select a column, we use temperature['column'].

We extend this knowledge by re-selecting (or subsetting) the dataset, but instead of writing the column that we want, we write the 'condition'.

In this case, the condition is that we want the 'city' column to only contains Moscow and St. Petersburg. 

```python
# Subset temperatures using square brackets
temperatures[temperatures['city'].isin(cities)]
```

Let's break it down one-by-one!

Try running the following.

```python
temperatures['city'].isin(cities)
```

Notice that it only shows either False or True. This is actually how pandas evaluate the expression. 

We select the column 'city', and we check if the cities are Moscow or St. Petersburg (by using isin). 

If the city is Moscow or St. Petersburg, then pandas will write down True. Otherwise, False.

Now, if we send this True/False column with a square bracket to temperatures, pandas will automatically select all rows that are True!

It is also possible to use other simpler expression as the following shows.

```python
temperatures[temperatures['city'] == 'Moscow']
```

Another way to subset the data is by imply setting the index and use .loc as usual.

```python
# Index temperatures by city
temperatures_ind = temperatures.set_index("city")

# Subset temperatures_ind using .loc[]
temperatures_ind.loc[cities]
```

## 3.3 Setting multi-level indexes
Indexes can also be made out of multiple columns, forming a multi-level index (sometimes called a hierarchical index). There is a trade-off to using these. <br/>
<br/>
The benefit is that multi-level indexes make it more natural to reason about nested categorical variables. For example, in a clinical trial, you might have control and treatment groups. Then each test subject belongs to one or another group, and we can say that a test subject is nested inside the treatment group. Similarly, in the temperature dataset, the city is located in the country, so we can say a city is nested inside the country.<br/>
<br/>
The main downside is that the code for manipulating indexes is different from the code for manipulating columns, so you have to learn two syntaxes and keep track of how your data is represented.<br/>

```python
# Index temperatures by country & city
temperatures_ind = temperatures.set_index(["country","city"])

# List of tuples: Brazil, Rio De Janeiro & Pakistan, Lahore
rows_to_keep = [("Brazil","Rio De Janeiro"),("Pakistan","Lahore")]

# Subset for rows to keep
temperatures_ind.loc[rows_to_keep]
```

You can also sort the index! Again, this is very useful when we are dealing with ordinal or time series data.

```python
# Sort temperatures_ind by index values
temperatures_ind.sort_index()
```

```python
# Sort temperatures_ind by index values at the city level
temperatures_ind.sort_index(level="city")
```

You can control the sorting, either asceding or descending, per index level.

```python
# Sort temperatures_ind by country then descending city
temperatures_ind.sort_index(level=["country","city"],ascending=[True,False])
```

## 3.4 Slicing index values
Slicing lets you select consecutive elements of an object using first:last syntax. DataFrames can be sliced by index values or by row/column number; 
<br/> we'll start with the first case. This involves slicing inside the .loc[] method.
<br/>
<br/>
Compared to slicing lists, there are a few things to remember.
<br/>
* You can only slice an index if the index is sorted (using .sort_index()).
* To slice at the outer level, first and last can be strings.
* To slice at inner levels, first and last should be tuples.
* If you pass a single slice to .loc[], it will slice the rows.

```python
# Sort the index of temperatures_ind
temperatures_srt = temperatures_ind.sort_index()

# Subset rows from Pakistan to Russia
temperatures_srt.loc["Pakistan":"Russia"]
```

```python
# Try to subset rows from Lahore to Moscow
temperatures_srt.loc["Lahore":"Moscow"]
```

```python
# Subset rows from Pakistan, Lahore to Russia, Moscow
temperatures_srt.loc[("Pakistan","Lahore"):("Russia","Moscow")]
```
